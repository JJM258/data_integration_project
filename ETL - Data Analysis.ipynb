# Data Analysis

## 1. Top 10 Traffic Disruptions Causes

```python
%sql
SELECT
    motivo,
    COUNT(*) AS total_disruptions
FROM etl_project.etl_schema.df_analysis_table
GROUP BY motivo
ORDER BY total_disruptions DESC
LIMIT 10;
```

## 2. Top 10 Most Common Impacts

```python
%sql
SELECT
    impacto,
    COUNT(*) AS impact_relevancy
FROM etl_project.etl_schema.df_analysis_table
GROUP BY impacto
ORDER BY impact_relevancy DESC;
```

## 3. Top 10 Most Effect Streets

```python
%sql
SELECT
    morada,
    COUNT(*) AS times_effected
FROM etl_project.etl_schema.df_analysis_table
GROUP BY morada
ORDER BY times_effected DESC
LIMIT 10;
```

## 4. Top Restriction Causes

```python
%sql
SELECT
    restricao_circulacao,
    COUNT(*) AS times_restricted
FROM etl_project.etl_schema.df_analysis_table
GROUP BY restricao_circulacao
ORDER BY times_restricted DESC;
```

## 5. Top 10 Comparision between its Impact and Restriction

```python
%sql
SELECT
    impacto,
    restricao_circulacao,
    COUNT(*) AS relevancy_and_restriction
FROM etl_project.etl_schema.df_analysis_table
GROUP BY impacto, restricao_circulacao
ORDER BY relevancy_and_restriction DESC
LIMIT 10;
```
